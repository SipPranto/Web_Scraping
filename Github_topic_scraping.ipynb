```python
from bs4 import BeautifulSoup
import requests
```

```python
topics_url='https://github.com/topics'
```

```python
response=requests.get(topics_url)
```

```python
response.status_code
```

```python
page_contents=response.text
```

```python
doc=BeautifulSoup(page_contents,'html.parser')
```

```python
blocks=doc.find_all('div',class_="py-4 border-bottom d-flex flex-justify-between")
```

```python
data=pd.DataFrame(columns=['topic','description','topic_url','repo_num'])
```

```python

base_url='https://github.com'
import requests

for block in blocks[0:4]:
  info=[]
  topic=block.find('p',class_="f3 lh-condensed mb-0 mt-1 Link--primary").text
  info.append(topic)
  description=block.find('p',class_="f5 color-fg-muted mb-0 mt-1").text.strip()
  info.append(description)
  topic_url=base_url+block.find('a',class_="no-underline flex-grow-0")['href']
  info.append(topic_url)

  #topic url page
  response1=requests.get(topic_url)
  page_content=response1.text

  page_doc=BeautifulSoup(page_content,'html.parser')
  repo_num=page_doc.find('h2',class_="h3 color-fg-muted").text.strip().replace('\n',' ').replace('  ','').split(' ')[2]
  info.append(repo_num)


  data.loc[len(data.index)] = info
```
