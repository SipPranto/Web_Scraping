```python
critics_certified = data.get('grid').get('list')[20].get('criticsScore').get('certified')
critics_score = data.get('grid').get('list')[20].get('criticsScore').get('scorePercent')
critics_sentiment = data.get('grid').get('list')[20].get('criticsScore').get('sentiment')
```

```python
list_ = []
for i in range(30):
    try:
        title = data.get('grid', {}).get('list', [])[i].get('title', None)
        type_ = data.get('grid', {}).get('list', [])[i].get('type', None)
        release_date = data.get('grid', {}).get('list', [])[i].get('releaseDateText', None)
        posterUrl = data.get('grid', {}).get('list', [])[i].get('posterUri', None)

        # Audience Score details with exception handling
        audience_certified = data.get('grid', {}).get('list', [])[i].get('audienceScore', {}).get('certified', None)
        audience_score = data.get('grid', {}).get('list', [])[i].get('audienceScore', {}).get('scorePercent', None)
        audience_sentiment = data.get('grid', {}).get('list', [])[i].get('audienceScore', {}).get('sentiment', None)

        # Critics Score details with exception handling
        critics_certified = data.get('grid', {}).get('list', [])[i].get('criticsScore', {}).get('certified', None)
        critics_score = data.get('grid', {}).get('list', [])[i].get('criticsScore', {}).get('scorePercent', None)
        critics_sentiment = data.get('grid', {}).get('list', [])[i].get('criticsScore', {}).get('sentiment', None)

        result = {
            'title': title,
            'type_': type_,
            'release_date': release_date,
            'posterUrl': posterUrl,
            'audience_certified': audience_certified,
            'audience_score': audience_score,
            'audience_sentiment': audience_sentiment,
            'critics_certified': critics_certified,
            'critics_score': critics_score,
            'critics_sentiment': critics_sentiment,
        }
        print(result)
        list_.append(result)

    except (IndexError, AttributeError, TypeError) as e:
        print(f"An error occurred while processing item {i}: {e}")
```

```python
df = pd.DataFrame(list_)
df.head(5)
```

```python

df['critics_score_val'] = df['critics_score'].apply(lambda x: (''.join(list(x)[:-1])).strip() if x is not '' else 0).astype(int)
df['audience_score_val'] = df['audience_score'].apply(lambda x: (''.join(list(x)[:-1])).strip() if x is not '' else 0).astype(int)
df['total_score']=df['audience_score_val']+df['critics_score_val']
```

```python
df.head(4)
```

# Top score movie

```python
df[df['total_score']==df['total_score'].max()]['title'].values[0]
```

# Movies total score above average

```python
df[df['total_score']>df['total_score'].mean()].sort_values(by='total_score',ascending=False).reset_index()['title']
```

```python
df['release_year']=df['release_date'].apply(lambda x: x.split(',')[-1].strip())
df['release_month']=df['release_date'].apply(lambda x: x.split(',')[0].split(' ')[1].strip())
df['release_day']=df['release_date'].apply(lambda x: x.split(',')[0].split(' ')[-1].strip())
df.head()
```

# Year wise average Score

```python
df.groupby('release_year')['total_score'].mean().sort_values(ascending=False).to_frame('Total Average Score')
```

```python
df.groupby('release_year')['total_score'].mean().sort_values(ascending=False).plot(kind='bar',figsize=( 10,6),ylabel='Average Rating',title='Yaer Wise Average Rating',color='tomato')
```

# top 3 movies for each year

```python
df['year_wise_rank']=df.groupby('release_year')['total_score'].rank(method='dense')
df[df['year_wise_rank']<4].sort_values(by=['release_year','year_wise_rank']).reset_index()[['release_year','title','year_wise_rank']]
```
